<b>Exercise 5. Display the <i>"food"</i> which has the highest number of orders?

```python
hightest_order = order_list['food'].value_counts().index[0]
print("The most ordered food item is:", hightest_order)
```

<b>Exercise 5. How many items were ordered for the most-ordered item?

```python
most_order = order_list['food'].value_counts().iloc[0]
print("The highest number of orders for a single food item is:", most_order)
```

<b>Exercise 6. Display the <i>"description"</i> which has the highest number of orders?</b>

```python
hightest_dec = order_list['description'].value_counts().index[0]
print("The most ordered food item is:", hightest_dec)
```


<b>Exercise 7. How many foods ordered in total?</b>

```python
qnt = order_list['quantity'].sum()
print("The total quantity of food items ordered is:", qnt)
```

<b>Exercise 8. Calculate the total income for the period in the dataset?</b>

```python
order_list['unit_price'] = (
    order_list['unit_price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .astype(float)
)
total_income = (order_list['quantity'] * order_list['unit_price']).sum()
print("The total income from food items ordered is:", total_income, "$")
```

<b>Exercise 9. Find the total number of orders.</b>

```python
total_orders = order_list['food'].value_counts().sum()
print("The total number of food orders is:", total_orders)
```

<b>Exercise 10. Find the average income for each order?</b>

```python
avg = total_income / total_orders
print("The average income per order is:", avg.round(2), "$")
```

<b>Exercise 11. How many different foods are ordered?</b>

```python
diff = order_list['food'].nunique()
print("The number of different food items ordered is:", diff)
```
